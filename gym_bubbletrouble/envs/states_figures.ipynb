```python
import gym
import sys
sys.path.append('../../bubbletrouble')
import bubbletrouble_env
from bubbletrouble_env import BubbleTroubleEnv
import matplotlib.pyplot as plt
import random

env = BubbleTroubleEnv(K=5, rand=True, timed=True)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20,10))
env.reset()
env.step(3)
current_frame = env.render()
previous_frame = env.render()
delta = current_frame - previous_frame

for i in range(4):
    for _ in range(20):
        state, *_ = env.step(random.randint(0,3))
        
    previous_frame = current_frame
    current_frame = env.render()
    
    delta = current_frame - previous_frame
    image = env.render_with_states()
    
    axes[0][i].imshow(image)
    axes[1][i].imshow(delta)
```
